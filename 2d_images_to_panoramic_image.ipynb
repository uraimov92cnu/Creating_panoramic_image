```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image1 = mpimg.imread("../datasets/panoramic images/4.jpg")
plt.imshow(image1)
plt.show()
image2 = mpimg.imread("../datasets/panoramic images/5.jpg")
plt.imshow(image2)
plt.show()
image3 = mpimg.imread("../datasets/panoramic images/6.jpg")
plt.imshow(image3)
plt.show()
image4 = mpimg.imread("../datasets/panoramic images/6.jpg")
plt.imshow(image4)
plt.show()
image5 = mpimg.imread("../datasets/panoramic images/7.jpg")
plt.imshow(image5)
plt.show()
images = [image1,image2,image3,image4,image5]
```
